```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs = 200
t = np.linspace(0, 10, Fs * 10)
f = 1
A =  0.075
y = A * np.sin(2*np.pi*f*t)
```

```python
plt.plot(t, y)
plt.xlabel('Time (s)')
plt.ylabel('Displacement (m)')
plt.show()
```

```python
v = np.gradient(y, t)
a = np.gradient(v, t)
```

```python
plt.plot(t, a)
plt.xlabel('Time (s)')
plt.ylabel('Accel (m/s2)')
plt.show()
```
